```julia
function convex_hull(C::Affine, D::Affine)
    
    vertices = Vec{2, Float64}[]
    
    l = max( length(C.γ), length(D.γ) )
    
    for ɛ in Iterators.product( ((-1, 1) for i in 1:l)... )
        
        x = C.c + sum(ɛ[i] * C.γ[i] for i in 1:length(C.γ))
        y = C.c + sum(ɛ[i] * D.γ[i] for i in 1:length(D.γ))
                                    
        push!(vertices, Vec(x, y))
    end 
                                
    hull = QuickHull.qhull(vertices)
    push!(hull, hull[1])

    xs = [vertex[1] for vertex in hull]
    ys = [vertex[2] for vertex in hull]
                                
    return xs, ys
end
```

```julia
A = 0.5 * [1 2; -1 1]

X = Affine(-1..1)
Y = Affine(-1..1)

XX = [X, Y]  # vect

XX = A * XX

xs, ys = convex_hull(XX[1], XX[2])
```

```julia
plot(xs, ys, fill=true, alpha=0.3, aspect_ratio=:equal)
```
